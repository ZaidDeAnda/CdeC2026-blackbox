### Collection data types

variables allow us to save single information in a useful way. But what about if we want to save 10 numbers, we have to write 10 variables? and what about 100?

luckily, we have **data collections**.

The data collection we'll be using are called **lists**

A list (or known as an array in other languages) is an **ordered** data structure, that means that the first data goes in the first spot (memory spot, more specifically), the second on the second, and so on. Every data has a sequential spotñ

They are declared this way:

```python
my_list = [value1, value2, value3]
```

You only have to put between brackets the values that you want to store
